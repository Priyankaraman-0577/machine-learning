```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/netflix_titles.csv")
print("Shape:", df.shape)
df.head()
```

```python
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(6,6))
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#3266ad','#73726c'])
plt.title('Movies vs TV Shows')
plt.ylabel('')
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')
df['year_added'] = df['date_added'].dt.year

plt.figure(figsize=(12,5))
df['year_added'].value_counts().sort_index().plot(kind='bar', color='#3266ad')
plt.title('Content Added Per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
df['country'].value_counts().head(10).plot(kind='bar', color='#534AB7')
plt.title('Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
genres = df['listed_in'].str.split(', ').explode()

plt.figure(figsize=(12,6))
genres.value_counts().head(10).plot(kind='barh', color='#1D9E75')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
df['rating'].value_counts().plot(kind='bar', color='#D85A30')
plt.title('Content Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
directors = df['director'].dropna()

plt.figure(figsize=(12,5))
directors.value_counts().head(10).plot(kind='bar', color='#BA7517')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Director')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("=== NETFLIX EDA INSIGHTS ===")
print(f"Total Titles: {df.shape[0]}")
print(f"Movies: {df[df['type']=='Movie'].shape[0]}")
print(f"TV Shows: {df[df['type']=='TV Show'].shape[0]}")
print(f"Top Country: {df['country'].value_counts().index[0]}")
print(f"Top Genre: {genres.value_counts().index[0]}")
print(f"Most Common Rating: {df['rating'].value_counts().index[0]}")
```

```python
import gradio as gr
import matplotlib.pyplot as plt
import io

def generate_netflix_eda():
    # Generate summary insights
    summary_text = "=== NETFLIX EDA INSIGHTS ===\n"
    summary_text += f"Total Titles: {df.shape[0]}\n"
    summary_text += f"Movies: {df[df['type']=='Movie'].shape[0]}\n"
    summary_text += f"TV Shows: {df[df['type']=='TV Show'].shape[0]}\n"
    summary_text += f"Top Country: {df['country'].value_counts().index[0]}\n"
    summary_text += f"Top Genre: {genres.value_counts().index[0]}\n"
    summary_text += f"Most Common Rating: {df['rating'].value_counts().index[0]}\n"

    # Plot 1: Movies vs TV Shows
    fig1, ax1 = plt.subplots(figsize=(6,6))
    df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#3266ad','#73726c'], ax=ax1)
    ax1.set_title('Movies vs TV Shows')
    ax1.set_ylabel('')
    buf1 = io.BytesIO()
    fig1.savefig(buf1, format='png')
    buf1.seek(0)
    plt.close(fig1)

    # Plot 2: Content Added Per Year
    fig2, ax2 = plt.subplots(figsize=(12,5))
    df['year_added'].value_counts().sort_index().plot(kind='bar', color='#3266ad', ax=ax2)
    ax2.set_title('Content Added Per Year')
    ax2.set_xlabel('Year')
    ax2.set_ylabel('Count')
    fig2.tight_layout()
    buf2 = io.BytesIO()
    fig2.savefig(buf2, format='png')
    buf2.seek(0)
    plt.close(fig2)

    return summary_text, buf1.getvalue(), buf2.getvalue()

# Create Gradio interface
iface = gr.Interface(
    fn=generate_netflix_eda,
    inputs=None,
    outputs=[
        gr.Textbox(label="EDA Insights"),
        gr.Image(label="Movies vs TV Shows Distribution"),
        gr.Image(label="Content Added Per Year")
    ],
    title="Netflix Content Exploratory Data Analysis",
    description="Explore key insights and distributions of Netflix content."
)

# Launch the interface
iface.launch()
```
